````python
import pandas as pd
````

````python
df = pd.read_csv("satRgbFinal4.csv")
````

````python
ref_point_keys = []
lats = []
lons = []
for label in df.sat_label:
    temp = df.loc[df.sat_label == label]
    lat_NW = temp['sat_Lat-NW'].values[0]
    lon_NW = temp['sat_Lon-NW'].values[0] 
    lat_SE = temp['sat_Lat-SE'].values[0]
    lon_SE = temp['sat_Lon-SE'].values[0] 
    lat_center = temp['sat_Lat-Center'].values[0]
    lon_center = temp['sat_Lon-Center'].values[0]
    for i in range(1, 4):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_NW)
        if i == 1:
            lons.append(lon_NW)
        elif i == 2:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
    for i in range(4, 7):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_center)
        if i == 4:
            lons.append(lon_NW)
        elif i == 5:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
    for i in range(7, 10):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_SE)
        if i == 7:
            lons.append(lon_NW)
        elif i == 8:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
````

````python
ref_lat_lons = pd.DataFrame(data = {'land_ref_label': ref_point_keys, 'land_Lat-Center': lats, 'land_Lon-Center': lons})
ref_lat_lons
````

Add Corners

````python

# Convert distance to degrees (approx)
LAT_BOUNDING_RADIUS_DEG = 69.0 # 69 miles per degree at 38 degrees North Latitude
LON_BOUNDING_RADIUS_DEG = 54.6 # 54.6 miles per degree at 38 degrees North Latitude for longitudinal directions
````

````python
def get_corners(lat, lon, radius):
    cover_lat_SE = lat - (radius/LAT_BOUNDING_RADIUS_DEG)
    cover_lat_NW = lat + (radius/LAT_BOUNDING_RADIUS_DEG)
    cover_lon_NW = lon - (radius/LON_BOUNDING_RADIUS_DEG)
    cover_lon_SE = lon + (radius/LON_BOUNDING_RADIUS_DEG)

    return [cover_lat_NW, cover_lon_NW, cover_lat_SE, cover_lon_SE]
````

````python
for ref_label in ref_lat_lons.land_ref_label:
    temp = ref_lat_lons.loc[ref_lat_lons.land_ref_label == ref_label]
    ref_corners = get_corners(temp['land_Lat-Center'].values[0], temp['land_Lon-Center'].values[0], .25)
    ref_lat_lons.loc[ref_lat_lons.land_ref_label == ref_label, 'land_Lat-NW'] = ref_corners[0]
    ref_lat_lons.loc[ref_lat_lons.land_ref_label == ref_label, 'land_Lon-NW'] = ref_corners[1]
    ref_lat_lons.loc[ref_lat_lons.land_ref_label == ref_label, 'land_Lat-SE'] = ref_corners[2]
    ref_lat_lons.loc[ref_lat_lons.land_ref_label == ref_label, 'land_Lon-SE'] = ref_corners[3]
````
