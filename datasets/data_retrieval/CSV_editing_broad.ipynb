```python
import pandas as pd
```

```python
df = pd.read_csv("../satelliteData/satRgbFinal4.csv")
```

```python
ref_point_keys = []
lats = []
lons = []
for label in df.sat_label:
    temp = df.loc[df.sat_label == label]
    lat_NW = temp['sat_Lat-NW'].values[0]
    lon_NW = temp['sat_Lon-NW'].values[0] 
    lat_SE = temp['sat_Lat-SE'].values[0]
    lon_SE = temp['sat_Lon-SE'].values[0] 
    lat_center = temp['sat_Lat-Center'].values[0]
    lon_center = temp['sat_Lon-Center'].values[0]
    for i in range(1, 4):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_NW)
        if i == 1:
            lons.append(lon_NW)
        elif i == 2:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
    for i in range(4, 7):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_center)
        if i == 4:
            lons.append(lon_NW)
        elif i == 5:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
    for i in range(7, 10):
        new_label = label + '_' + str(i)
        ref_point_keys.append(new_label)
        lats.append(lat_SE)
        if i == 7:
            lons.append(lon_NW)
        elif i == 8:
            lons.append(lon_center)
        else:
            lons.append(lon_SE)
```

```python
ref_lat_lons = pd.DataFrame(data = {'land_ref_label': ref_point_keys, 'land_Lat-Center': lats, 'land_Lon-Center': lons})
ref_lat_lons
```

Add empty region columns

```python
ref_lat_lons[['GRASSY', 'DENSE_WOOD', 'WOODY', 'SUBURBAN', 'WATERY', 'URBAN', 'AGRICULTURAL']] = 0
ref_lat_lons
```

Let's save this to CSV for collecting the landcover percentages.

```python
ref_lat_lons.tail(9).to_csv("broad_landcover.csv", index = False)
```

## Stitching together results

We'll pull in the main landcover file (`landcover-nj-2023.csv`) which already has local values to add in the broad values.

```python
df2 = pd.read_csv("../satelliteData/landcover-nj-2023.csv")
df2.head()
```

```python
df2.tail()
```

Read in majority of broad landcover data (missing Urban Library).

```python
broad_regions = pd.read_csv("broad_landcover_complete.csv")
broad_regions
```

Get the urban library broad data too.

```python
broad_lib = pd.read_csv("broad_landcover_library_complete.csv")
```

```python
broad_lib.head()
```

Add the first chunk, then we'll add the library data.

```python
for ref_label in df2.land_ref_label[:162]:
    temp = broad_regions.loc[broad_regions.ref_label == ref_label]
    df2.loc[df2.land_ref_label == ref_label, 'GRASSY_BROAD'] = temp.GRASSY
    df2.loc[df2.land_ref_label == ref_label, 'DENSE_WOOD_BROAD'] = temp.DENSE_WOOD
    df2.loc[df2.land_ref_label == ref_label, 'WOODY_BROAD'] = temp.WOODY
    df2.loc[df2.land_ref_label == ref_label, 'SUBURBAN_BROAD'] = temp.SUBURBAN
    df2.loc[df2.land_ref_label == ref_label, 'WATERY_BROAD'] = temp.WATERY
    df2.loc[df2.land_ref_label == ref_label, 'URBAN_BROAD'] = temp.URBAN
    df2.loc[df2.land_ref_label == ref_label, 'AGRICULTURAL_BROAD'] = temp.AGRICULTURAL
```

```python
for ref_label in df2.land_ref_label[162:]:
    temp = broad_lib.loc[broad_lib.ref_label == ref_label]
    df2.loc[df2.land_ref_label == ref_label, 'GRASSY_BROAD'] = temp.GRASSY.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'DENSE_WOOD_BROAD'] = temp.DENSE_WOOD.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'WOODY_BROAD'] = temp.WOODY.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'SUBURBAN_BROAD'] = temp.SUBURBAN.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'WATERY_BROAD'] = temp.WATERY.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'URBAN_BROAD'] = temp.URBAN.values[0]
    df2.loc[df2.land_ref_label == ref_label, 'AGRICULTURAL_BROAD'] = temp.AGRICULTURAL.values[0]
```
